```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

# Define paths
data_dir = '/content/aptos2019'
base_dir = os.path.join(data_dir, 'APTOS 2019 (Original) (Binary)')
output_dir = base_dir  # Save CSVs where training code expects them

# Function to collect images and labels
def collect_images_and_labels(base_dir):
    image_files = []
    labels = []

    # Possible class folder names for No DR and DR
    no_dr_folders = ['0', 'No_DR', 'NoDR', 'no_dr', 'nodr']
    dr_folders = ['1', 'DR', 'dr']

    # Check for class subfolders
    for subdir in os.listdir(base_dir):
        subdir_path = os.path.join(base_dir, subdir)
        if os.path.isdir(subdir_path):
            for f in os.listdir(subdir_path):
                if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                    # Store relative path (e.g., 'DR/image.jpg')
                    rel_path = os.path.join(subdir, f)
                    image_files.append(rel_path)
                    # Assign label: 0 for No DR, 1 for DR
                    if subdir.lower() in [x.lower() for x in no_dr_folders]:
                        labels.append(0)
                    elif subdir.lower() in [x.lower() for x in dr_folders]:
                        labels.append(1)
                    else:
                        labels.append(0)  # Default to No DR if folder name unclear

    # If no class subfolders, try direct images in base_dir
    if not image_files:
        for f in os.listdir(base_dir):
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                rel_path = f
                image_files.append(rel_path)
                labels.append(0)  # Placeholder: No DR

    # Try one level deeper (e.g., 'train_images')
    if not image_files:
        for subdir in os.listdir(base_dir):
            subdir_path = os.path.join(base_dir, subdir)
            if os.path.isdir(subdir_path):
                for f in os.listdir(subdir_path):
                    if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                        rel_path = os.path.join(subdir, f)
                        image_files.append(rel_path)
                        # Assign label based on subdir
                        if subdir.lower() in [x.lower() for x in no_dr_folders]:
                            labels.append(0)
                        elif subdir.lower() in [x.lower() for x in dr_folders]:
                            labels.append(1)
                        else:
                            labels.append(0)  # Default to No DR

    if not image_files:
        raise FileNotFoundError(f"No images found in {base_dir} or its subfolders. Run '!ls -R {data_dir}' to check the structure.")

    return image_files, labels

# Collect images and labels
image_files, labels = collect_images_and_labels(base_dir)

# Create DataFrame
df = pd.DataFrame({
    'id_code': image_files,
    'diagnosis': labels
})

# Split the dataset: 70% train, 15% validation, 15% test
train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df['diagnosis'])
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['diagnosis'])

# Save CSV files
train_df.to_csv(os.path.join(output_dir, 'train.csv'), index=False)
val_df.to_csv(os.path.join(output_dir, 'validation.csv'), index=False)
test_df.to_csv(os.path.join(output_dir, 'test.csv'), index=False)

print("CSV files created successfully in", output_dir)
print(f"Train set size: {len(train_df)}")
print(f"Validation set size: {len(val_df)}")
print(f"Test set size: {len(test_df)}")
print("\nLabel distribution in train.csv:")
print(train_df['diagnosis'].value_counts())
print("\nSample from test.csv:")
print(pd.read_csv(os.path.join(output_dir, 'test.csv')).head())
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.metrics import Precision, Recall
from tensorflow.keras.utils import Sequence
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import pickle

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Configuration
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 50
LEARNING_RATE = 0.0001

# Define paths
BASE_DIR = '/content/aptos2019/APTOS 2019 (Original) (Binary)'
TRAIN_CSV = os.path.join(BASE_DIR, 'train.csv')
VALIDATION_CSV = os.path.join(BASE_DIR, 'validation.csv')
TEST_CSV = os.path.join(BASE_DIR, 'test.csv')

# Generate CSV files
print("Generating CSV files...")
def generate_csv_files(base_dir):
    image_files = []
    labels = []
    no_dr_folders = ['0', 'No_DR', 'NoDR', 'no_dr', 'nodr']
    dr_folders = ['1', 'DR', 'dr']

    # Check for class subfolders
    for subdir in os.listdir(base_dir):
        subdir_path = os.path.join(base_dir, subdir)
        if os.path.isdir(subdir_path):
            for f in os.listdir(subdir_path):
                if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                    rel_path = os.path.join(subdir, f)
                    image_files.append(rel_path)
                    if subdir.lower() in [x.lower() for x in no_dr_folders]:
                        labels.append(0)  # No DR
                    elif subdir.lower() in [x.lower() for x in dr_folders]:
                        labels.append(1)  # DR
                    else:
                        labels.append(0)

    # Try direct images in base_dir
    if not image_files:
        for f in os.listdir(base_dir):
            if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                rel_path = f
                image_files.append(rel_path)
                labels.append(0)

    # Try one level deeper
    if not image_files:
        for subdir in os.listdir(base_dir):
            subdir_path = os.path.join(base_dir, subdir)
            if os.path.isdir(subdir_path):
                for f in os.listdir(subdir_path):
                    if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                        rel_path = os.path.join(subdir, f)
                        image_files.append(rel_path)
                        if subdir.lower() in [x.lower() for x in no_dr_folders]:
                            labels.append(0)
                        elif subdir.lower() in [x.lower() for x in dr_folders]:
                            labels.append(1)
                        else:
                            labels.append(0)

    if not image_files:
        raise FileNotFoundError(f"No images found in {base_dir} or its subfolders. Run '!ls -R /content/aptos2019' to check the structure.")

    df = pd.DataFrame({'id_code': image_files, 'diagnosis': labels})
    train_df, temp_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df['diagnosis'])
    val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['diagnosis'])

    train_df.to_csv(TRAIN_CSV, index=False)
    val_df.to_csv(VALIDATION_CSV, index=False)
    test_df.to_csv(TEST_CSV, index=False)

    print(f"CSV files created in {BASE_DIR}")
    print(f"Train set size: {len(train_df)}")
    print(f"Validation set size: {len(val_df)}")
    print(f"Test set size: {len(test_df)}")
    print("\nLabel distribution in train.csv:")
    print(train_df['diagnosis'].value_counts())
    print("\nSample from test.csv:")
    print(pd.read_csv(TEST_CSV).head())

# Run CSV generation if CSVs don't exist
if not (os.path.exists(TRAIN_CSV) and os.path.exists(VALIDATION_CSV) and os.path.exists(TEST_CSV)):
    generate_csv_files(BASE_DIR)

# Explore dataset structure
print("\nExploring dataset structure...")
if os.path.exists(BASE_DIR):
    contents = os.listdir(BASE_DIR)
    print(f"Directory contents: {contents}")
    for item in contents:
        item_path = os.path.join(BASE_DIR, item)
        if os.path.isdir(item_path):
            num_files = len([f for f in os.listdir(item_path) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp'))])
            print(f"  Folder '{item}': {num_files} images")
        elif item.endswith('.csv'):
            print(f"  CSV file: {item}")
        else:
            print(f"  File: {item}")

# Check CSV files
if os.path.exists(TEST_CSV):
    print(f"\nExamining test.csv:")
    test_df = pd.read_csv(TEST_CSV)
    print(f"Shape: {test_df.shape}")
    print(f"Columns: {test_df.columns.tolist()}")
    print(f"First few rows:")
    print(test_df.head())
if os.path.exists(TRAIN_CSV):
    print(f"\nExamining train.csv:")
    train_df = pd.read_csv(TRAIN_CSV)
    print(f"Shape: {train_df.shape}")
    print(f"Columns: {train_df.columns.tolist()}")
    print(f"Label distribution:")
    print(train_df['diagnosis'].value_counts())

# Custom CSVDataGenerator as Keras Sequence
class CSVDataGenerator(Sequence):
    def __init__(self, csv_path, image_dir, target_size=(224, 224), batch_size=32,
                 class_mode='binary', shuffle=True, augment=False):
        self.csv_path = csv_path
        self.image_dir = image_dir
        self.target_size = target_size
        self.batch_size = batch_size
        self.class_mode = class_mode
        self.shuffle = shuffle
        self.augment = augment
        self.df = pd.read_csv(csv_path)
        self.samples = len(self.df)
        self.image_col = None
        self.label_col = None
        for col in self.df.columns:
            if 'image' in col.lower() or 'id' in col.lower() or col.lower() in ['filename', 'file']:
                self.image_col = col
            if 'diagnosis' in col.lower() or 'label' in col.lower() or 'target' in col.lower():
                self.label_col = col
        if self.image_col is None:
            self.image_col = self.df.columns[0]
        if self.label_col is None and len(self.df.columns) > 1:
            self.label_col = self.df.columns[1]
        print(f"Using '{self.image_col}' as image column")
        print(f"Using '{self.label_col}' as label column")
        if augment:
            self.datagen = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.2,
                height_shift_range=0.2,
                shear_range=0.2,
                zoom_range=0.2,
                horizontal_flip=True,
                fill_mode='nearest'
            )
        else:
            self.datagen = None
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(self.samples / self.batch_size))

    def __getitem__(self, idx):
        batch_start = idx * self.batch_size
        batch_end = min((idx + 1) * self.batch_size, self.samples)
        batch_images = []
        batch_labels = []
        for i in range(batch_start, batch_end):
            img_name = self.df.iloc[i][self.image_col]
            if not img_name.endswith(('.jpg', '.jpeg', '.png')):
                img_name += '.jpg'
            img_path = os.path.join(self.image_dir, img_name)
            try:
                img = load_img(img_path, target_size=self.target_size)
                img_array = img_to_array(img) / 255.0
                if self.datagen is not None:
                    img_array = self.datagen.random_transform(img_array)
                batch_images.append(img_array)
                if self.label_col:
                    label = self.df.iloc[i][self.label_col]
                    if isinstance(label, str):
                        label = 1 if label.lower() in ['dr', 'diabetic', 'positive', '1'] else 0
                    else:
                        label = int(label)
                    batch_labels.append(label)
                else:
                    batch_labels.append(0)
            except Exception as e:
                print(f"Error loading image {img_path}: {e}")
                dummy_img = np.zeros((*self.target_size, 3))
                batch_images.append(dummy_img)
                batch_labels.append(0)
        return np.array(batch_images), np.array(batch_labels)

    def on_epoch_end(self):
        if self.shuffle:
            self.df = self.df.sample(frac=1).reset_index(drop=True)

# Create data generators
print("\nSetting up CSV-based data generators...")
image_dir = BASE_DIR
if os.path.exists(TRAIN_CSV) and os.path.exists(VALIDATION_CSV):
    train_generator = CSVDataGenerator(
        TRAIN_CSV,
        image_dir,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        augment=True
    )
    validation_generator = CSVDataGenerator(
        VALIDATION_CSV,
        image_dir,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        augment=False
    )
    test_generator = None
    if os.path.exists(TEST_CSV):
        test_generator = CSVDataGenerator(
            TEST_CSV,
            image_dir,
            target_size=(IMG_SIZE, IMG_SIZE),
            batch_size=BATCH_SIZE,
            augment=False,
            shuffle=False
        )
    steps_per_epoch = train_generator.samples // BATCH_SIZE
    validation_steps = validation_generator.samples // BATCH_SIZE
    test_steps = test_generator.samples // BATCH_SIZE if test_generator else None
    print("CSV-based data generators created successfully!")
else:
    raise FileNotFoundError(f"Missing CSV files: {TRAIN_CSV}, {VALIDATION_CSV}")

# Build ResNet50 Model
def create_resnet50_model():
    base_model = ResNet50(
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base_model.trainable = False
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.3)(x)
    predictions = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    return model

# Create and compile model
model = create_resnet50_model()
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(name='precision'), Recall(name='recall')]
)

print("\nModel summary:")
model.summary()

# Callbacks
callbacks = [
    ModelCheckpoint(
        'best_resnet50_dr_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        save_weights_only=False,
        verbose=1
    ),
    EarlyStopping(
        monitor='val_loss',
        patience=10,
        restore_best_weights=True,
        verbose=1
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    )
]

# Train the model
print("\nStarting training...")
history = model.fit(
    train_generator,
    steps_per_epoch=steps_per_epoch,
    epochs=EPOCHS,
    validation_data=validation_generator,
    validation_steps=validation_steps,
    callbacks=callbacks,
    verbose=1
)

# Plot training history
def plot_training_history(history):
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    axes[0, 0].plot(history.history['accuracy'], label='Training Accuracy')
    axes[0, 0].plot(history.history['val_accuracy'], label='Validation Accuracy')
    axes[0, 0].set_title('Model Accuracy')
    axes[0, 0].set_xlabel('Epoch')
    axes[0, 0].set_ylabel('Accuracy')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    axes[0, 1].plot(history.history['loss'], label='Training Loss')
    axes[0, 1].plot(history.history['val_loss'], label='Validation Loss')
    axes[0, 1].set_title('Model Loss')
    axes[0, 1].set_xlabel('Epoch')
    axes[0, 1].set_ylabel('Loss')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    axes[1, 0].plot(history.history['precision'], label='Training Precision')
    axes[1, 0].plot(history.history['val_precision'], label='Validation Precision')
    axes[1, 0].set_title('Model Precision')
    axes[1, 0].set_xlabel('Epoch')
    axes[1, 0].set_ylabel('Precision')
    axes[1, 0].legend()
    axes[1, 0].grid(True)
    axes[1, 1].plot(history.history['recall'], label='Training Recall')
    axes[1, 1].plot(history.history['val_recall'], label='Validation Recall')
    axes[1, 1].set_title('Model Recall')
    axes[1, 1].set_xlabel('Epoch')
    axes[1, 1].set_ylabel('Recall')
    axes[1, 1].legend()
    axes[1, 1].grid(True)
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()

plot_training_history(history)

# Load the best model
model = tf.keras.models.load_model('best_resnet50_dr_model.h5')

# Evaluation
print("\n" + "="*50)
print("MODEL EVALUATION")
print("="*50)

# Evaluate on validation set
val_loss, val_accuracy, val_precision, val_recall = model.evaluate(
    validation_generator,
    steps=validation_steps,
    verbose=1
)
print(f"\nValidation Results:")
print(f"Loss: {val_loss:.4f}")
print(f"Accuracy: {val_accuracy:.4f}")
print(f"Precision: {val_precision:.4f}")
print(f"Recall: {val_recall:.4f}")
print(f"F1-Score: {2 * (val_precision * val_recall) / (val_precision + val_recall + 1e-7):.4f}")

# Generate predictions
validation_generator.reset()
predictions = model.predict(validation_generator, steps=validation_steps, verbose=1)
predicted_classes = (predictions > 0.5).astype(int)

# Get true labels
true_classes = []
for i in range(validation_steps):
    _, labels = next(validation_generator)
    true_classes.extend(labels)
true_classes = np.array(true_classes)[:len(predicted_classes)]

# Classification report
class_names = ['No DR', 'DR']
print("\nClassification Report:")
print(classification_report(true_classes, predicted_classes, target_names=class_names))

# Confusion Matrix
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('confusion_matrix.png')
plt.show()

# Test set evaluation
if test_generator:
    print("\n" + "="*50)
    print("TEST SET EVALUATION")
    print("="*50)
    test_loss, test_accuracy, test_precision, test_recall = model.evaluate(
        test_generator,
        steps=test_steps,
        verbose=1
    )
    print(f"\nTest Results:")
    print(f"Loss: {test_loss:.4f}")
    print(f"Accuracy: {test_accuracy:.4f}")
    print(f"Precision: {test_precision:.4f}")
    print(f"Recall: {test_recall:.4f}")
    print(f"F1-Score: {2 * (test_precision * test_recall) / (test_precision + test_recall + 1e-7):.4f}")

# Predict function
def predict_image(image_path):
    img = load_img(image_path, target_size=(IMG_SIZE, IMG_SIZE))
    img_array = img_to_array(img)
    img_array = tf.expand_dims(img_array, 0) / 255.0
    prediction = model.predict(img_array)[0][0]
    result = "Diabetic Retinopathy Detected" if prediction > 0.5 else "No Diabetic Retinopathy"
    confidence = prediction if prediction > 0.5 else 1 - prediction
    return result, confidence

print("\n" + "="*50)
print("MODEL TRAINING COMPLETED!")
print("="*50)
print(f"Best model saved as: best_resnet50_dr_model.h5")
print("Training history saved as: training_history.pkl")
with open('training_history.pkl', 'wb') as f:
    pickle.dump(history.history, f)
```
